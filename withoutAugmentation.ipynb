```
import numpy as np
from keras.models import Sequential
from keras.utils import to_categorical
from sklearn.model_selection import train_test_split
from google.colab import drive
import cv2
import matplotlib.pyplot as plt
import os
```

```
ALL_CARS_NAMES = os.listdir('drive/My Drive/PAK dataset/Final')
```

```
ALL_CARS_NAMES
```

LABELS GENERATIONS

```
CARS_LOCATION = 'drive/My Drive/PAK dataset/Final/'
Total_Y_labels = []
count = 0 #<== Label number of each count
for names in ALL_CARS_NAMES: #<== getting all car names one by one, 
  location = CARS_LOCATION+str(names) #<== getting the file location of the cars one by one
  all_single_cars_images = os.listdir(location) #<== getting all the images names from that file folder
  count_46 = 0
  for i in range(len(all_single_cars_images)): #<== iterating total number of images 
    if 'jpg' in all_single_cars_images[i] or 'jpeg' in all_single_cars_images[i] or 'png' in all_single_cars_images[i]:
      # if any images of jpg/png/jpeg format exists, the append 
      #print(all_single_cars_images[i])
      Total_Y_labels.append(count) #<== appending count in to ylabel, acc to total numbers of images
      count_46 +=1
      if count_46 == 46:
        break
  print("---------------------------")
  print('{} labels is ===>  {}'.format(names,count))
  count+=1 #add 1, when after appending all the labels of the specific cars, and move next to another car
```

X_INPUT GENERATION

```
all_images = []
CARS_LOCATION = 'drive/My Drive/PAK dataset/Final/'
for names in ALL_CARS_NAMES:
  location = CARS_LOCATION+str(names)
  all_images_of_the_car = os.listdir(CARS_LOCATION+str(names))
  count =0
  print("-----------------------------------")
  print(names)
  for name in all_images_of_the_car:
    if 'jpg' in name or 'png' in name or 'jpeg' in name:
      #print(location+'/'+str(name))
      img = cv2.imread(location+'/'+str(name))
      img = cv2.resize(img,(300,300))
      all_images.append(img)
      count += 1
      print(count)
      if count == 46:
        break
```

```
x = all_images
```

```
y = Total_Y_labels
```

```
x_input = np.array(x)
y_label = np.array(y)
```

```
xtrain,xtest,ytrain,ytest = train_test_split(x_input,y_label,test_size=0.30,random_state=7)
```

```
xtrain.shape,ytrain.shape,xtest.shape,ytest.shape
```

```
from keras.applications import ResNet50
```

```
pretrained_model = ResNet50(include_top=False, weights='imagenet', input_shape=(300,300,3))
```

```
from keras.utils import to_categorical
ytrain = to_categorical(ytrain)
ytest = to_categorical(ytest)
```

```
ytrain.shape,ytest.shape
```

```
from keras.layers import GlobalAveragePooling2D,Dense,Dropout,Flatten
models = Sequential()
models.add(pretrained_model)
models.add(Flatten())
models.add(Dense(256,activation='relu'))
models.add(Dropout(0.5))
models.add(Dense(100,activation='relu'))
models.add(Dropout(0.5))
models.add(Dense(37,activation='softmax'))
```

```
models.summary()
```

```
models.compile(optimizer='rmsprop',loss='categorical_crossentropy',metrics=['accuracy'])
```

```
models.fit(xtrain,ytrain,epochs=5,batch_size=120)
```
